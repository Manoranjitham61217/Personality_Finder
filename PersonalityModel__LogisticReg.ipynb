```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
le=LabelEncoder()
pf['personality']=le.fit_transform(pf['personality_type'])
```

```python
pf.head()
```

```python
x=pf.drop(['personality','personality_type'],axis=1)
y=pf['personality']
```

```python
print(x)
```

```python
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.scatter(pf['alone_time_preference'],y,color='red')
plt.title("Distribution of Personality Type by their Alone Time")
plt.xlabel("Alone Time")
plt.ylabel("Personality")
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
selector = SelectKBest(score_func=f_regression, k=10)
selector.fit(x, y)
selected_features = x.columns[selector.get_support()]
print("Selected Features:")
print(selected_features)
```

```python
x_selected=x[selected_features]
print(x_selected)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_selected)
```

```python
df_num = pf.select_dtypes(include='number')
corr_target = df_num.corr()['personality'].abs().sort_values(ascending=False)
```

```python
print(corr_target)
```

```python
top_features = corr_target[1:11].index
print(top_features)
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(df_num[top_features].corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Heatmap of Top 10 Features Correlated with Personality")
plt.show()
```

```python
top10 = corr_target[1:11] 
top10.plot(kind='barh', figsize=(8,6))
plt.xlabel("Correlation with Personality")
plt.title("Top  Features Correlated with Personality")
plt.show()
```

```python
print(x_scaled.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
c_matrix=confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",c_matrix)
```

```python
report=classification_report(y_test,y_pred)
print("Classification Report: \n",report)
```
